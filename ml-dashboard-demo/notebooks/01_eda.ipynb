# California Housing — Exploratory Data Analysis

A quick look at the dataset before modelling: feature distributions, the
target, and correlations.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import plotly.express as px

from backend.data import load_raw, add_engineered_features, TARGET

X, y = load_raw()
X = add_engineered_features(X)
df = X.assign(**{TARGET: y})
df.shape
```

```python
df.describe().T
```

## Target distribution

Median house value is capped at 5.0 ($500k), which shows up as a spike on the
right — worth remembering when reading the residuals.

```python
px.histogram(df, x=TARGET, nbins=50, title='Median house value ($100k)')
```

## Correlations with the target

```python
corr = df.corr(numeric_only=True)[TARGET].sort_values()
px.bar(corr.drop(TARGET), orientation='h', title='Correlation with median house value')
```

`MedInc` (median income) is by far the strongest signal — no surprise. The
engineered `rooms_per_person` adds a little extra. On to modelling in
`02_model_experiments.ipynb`.
